```python
ind49_rets = erk.get_ind_returns(weighting="vw", n_inds=49)["1974":]
ind49_rets.head()
```

# Assignment

```python
industry = "Steel"
```

```python
df_industry = ind49_rets[[industry]]
df_industry
```

```python
import statsmodels.api as sm
import numpy as np
ind49_excess = df_industry["2013":] - fff.loc["2013":, ['RF']].values
mkt_excess = fff.loc["2013":,['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Alpha"] = 1
lm = sm.OLS(ind49_excess, exp_var).fit()
```

```python
lm.summary()
```

```python
lm.params[0]
```

```python
industry_beta = {}
for industry in ind49_rets.columns:
    df_industry = ind49_rets[[industry]]
    ind49_excess = df_industry["1991":"1993"] - fff.loc["1991":"1993", ['RF']].values
    mkt_excess = fff.loc["1991":"1993":,['Mkt-RF']]
    exp_var = mkt_excess.copy()
    exp_var["Alpha"] = 1
    lm = sm.OLS(ind49_excess, exp_var).fit()

    industry_beta[industry] = lm.params[0]
industry_beta = pd.DataFrame(industry_beta, index=["Beta"]).T
```

```python
industry_beta.sort_values("Beta").iloc[[0,-1]]
```

```python
industry_beta = {}

for industry in ind49_rets.columns:
    df_industry = ind49_rets[[industry]]
    ind49_excess = df_industry["1991":] - fff.loc["1991":, ['RF']].values
    mkt_excess = fff.loc["1991"::,['Mkt-RF']]
    exp_var = mkt_excess.copy()
    exp_var["Value"] = fff.loc["1991":,['HML']]
    exp_var["Size"] = fff.loc["1991":,['SMB']]
    exp_var["Alpha"] = 1
    lm = sm.OLS(ind49_excess, exp_var).fit()

    industry_beta[industry] = {
        "beta":lm.params[0],
        "value":lm.params[1],
        "size":lm.params[2],
        "alpha":lm.params[3],
    }
industry_beta = pd.DataFrame(industry_beta,).T
```
